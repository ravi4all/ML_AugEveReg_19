```python
votes = [355,110,45,34]
parties = ['BJP','CONG','SP+BSP','Others']
```

```python
plt.figure(figsize=(10,6))
plt.pie(x=votes,labels=parties,autopct='%1.2f%%',startangle=90,counterclock=False,explode=(0,0,0.3,0))
plt.title("Election Result 2019")
plt.legend()
plt.show()
```

```python
age = np.random.randint(18,100,30)
```

```python
plt.hist(age,rwidth=0.9)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Frequency Data')
plt.show()
```
