```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('ipl/matches.csv')
deliveries = pd.read_csv('ipl/deliveries.csv')
```

```python
matches.shape
```

```python
deliveries.shape
```

```python
pd.value_counts(matches['season'])
```

```python
sns.countplot(x='season',data=matches)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='winner',data=matches)
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.countplot(x='toss_decision',data=matches)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='season',data=matches,hue='toss_decision')
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = matches['winner'].value_counts().plot.bar(width=0.9,color=sns.color_palette('inferno'))
for patch in ax.patches:
    ax.annotate(format(patch.get_height()),(patch.get_x() + 0.3,patch.get_height() + 1))
```

```python
plt.figure(figsize=(10,6))
ax = deliveries.groupby(['batsman','match_id'])['batsman_runs'].sum().sort_values(ascending=False).head(10).plot.bar(width=0.9,color=sns.color_palette('inferno'))
for patch in ax.patches:
    ax.annotate(format(patch.get_height()),(patch.get_x() + 0.3,patch.get_height() + 1))
```
